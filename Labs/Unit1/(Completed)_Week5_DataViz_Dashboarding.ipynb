### Q1 ‚Äî Exploratory vs. Explanatory
**Prompt:** *"Explain the difference between exploratory and explanatory visualization."*
**Hints:** Consider audience, purpose, level of polish, and how much iteration you expect.

üìù **Your response:**

*   **Exploratory Visualization:**
    *   **Audience:** Primarily for the data analyst or creator.
    *   **Purpose:** To understand the data, identify patterns, outliers, and test hypotheses. It's about discovery.
    *   **Level of polish:** Often rough and unpolished. Focus is on quick insights rather than aesthetics.
    *   **Iteration:** Highly iterative. You'll likely create many visualizations, discarding most.
*   **Explanatory Visualization:**
    *   **Audience:** For a specific audience (e.g., executives, clients, the public).
    *   **Purpose:** To communicate a specific finding, story, or insight clearly and effectively. It's about presentation.
    *   **Level of polish:** Highly polished and refined. Focus is on clarity, accuracy, and visual appeal.
    *   **Iteration:** Less iterative in terms of generating many charts; more iterative in refining a few key charts for maximum impact.

### Q2 ‚Äî Chart Types & When to Use Them
**Prompt:** *"Describe 5 common chart types (e.g., bar, line, scatter, histogram, map) and the type of data analysis question each is best suited to answer."*

For each chart, list: **Question archetype**, **Data shape**, **Pitfalls**, **Alternatives**.

üìù **Your response:**
1.  **Bar Chart:**
    *   **Question archetype:** How do values compare across different categories? (Comparison)
    *   **Data shape:** One categorical variable and one quantitative variable.
    *   **Pitfalls:** Too many categories can make it unreadable; not ideal for showing distributions.
    *   **Alternatives:** Column chart (vertical bar), Packed bubble chart, Treemap.
2.  **Line Chart:**
    *   **Question archetype:** How does a value change over time or a continuous variable? (Trend over time, Distribution over interval)
    *   **Data shape:** One temporal or continuous variable and one quantitative variable.
    *   **Pitfalls:** Can be misleading with irregular time intervals or if the y-axis doesn't start at zero (for some data).
    *   **Alternatives:** Area chart, Sparkline, Connected Scatter plot.
3.  **Scatter Plot:**
    *   **Question archetype:** Is there a relationship between two quantitative variables? (Relationship)
    *   **Data shape:** Two quantitative variables.
    *   **Pitfalls:** Overplotting with large datasets; difficulty seeing patterns if no strong correlation exists.
    *   **Alternatives:** Bubble chart (adds a third quantitative variable), Heatmap (for showing relationships between categories).
4.  **Histogram:**
    *   **Question archetype:** How is a quantitative variable distributed? (Distribution)
    *   **Data shape:** One quantitative variable.
    *   **Pitfalls:** The choice of bin size can significantly alter the appearance of the distribution; not good for comparing multiple distributions easily on the same chart.
    *   **Alternatives:** Density plot, Box plot, Violin plot.
5.  **Map:**
    *   **Question archetype:** How do values vary geographically? (Geographic)
    *   **Data shape:** Geographic data (latitude/longitude, region names) and one or more quantitative or categorical variables.
    *   **Pitfalls:** Can be misleading if area size is confused with the value; choosing appropriate color scales is critical.
    *   **Alternatives:** Symbol map, Choropleth map, Dot map.

### Q3 ‚Äî Airline Executive Dashboard KPIs
**Prompt:** *"Your goal is to create a dashboard for an airline executive. Brainstorm 5 Key Performance Indicators (KPIs) that should be on this dashboard. For each KPI, suggest the best chart type to visualize it."*

Suggested KPI ideas to consider (customize them): On-Time Rate, Average Delay Minutes, Cancellation Rate, Load Factor, Yield (Revenue per Passenger Mile), Completion Factor, Net Promoter Score (if available), Revenue, Cost per Available Seat Mile (CASM).

üìù **Your response:**
1.  **KPI:** On-Time Rate
    *   **Why it matters:** Directly impacts customer satisfaction and operational efficiency. High on-time rates indicate reliable service.
    *   **Best chart:** Gauge chart or a simple number card for the overall rate, with a time series chart (line or area) to show trends over time. A bar chart can show on-time rate by carrier or route.
    *   **Drill-down fields:** Carrier, Origin-Destination Route, Date/Month.
2.  **KPI:** Average Arrival Delay (Minutes)
    *   **Why it matters:** Another key indicator of operational performance and customer experience. Lower average delays are better.
    *   **Best chart:** Gauge chart or number card for the overall average, line chart for trend over time (perhaps with a rolling average), and a bar chart for average delay by carrier or route.
    *   **Drill-down fields:** Carrier, Origin-Destination Route, Date/Month, Time of Day.
3.  **KPI:** Load Factor (%)
    *   **Why it matters:** Measures how full the planes are, directly impacting revenue and profitability. Higher load factors are generally better, up to a point.
    *   **Best chart:** Gauge chart or number card for the overall average, line chart for trend over time, and a bar chart for load factor by route or aircraft type.
    *   **Drill-down fields:** Route, Aircraft Type, Date/Month, Carrier.
4.  **KPI:** Revenue (USD)
    *   **Why it matters:** The primary measure of financial performance. Executives need to track revenue generation.
    *   **Best chart:** Number card for total revenue, line chart for revenue trend over time, bar chart for revenue by carrier or route. Could also use a treemap to show revenue breakdown by route.
    *   **Drill-down fields:** Carrier, Route, Date/Month, Fare Class.
5.  **KPI:** Cost per Available Seat Mile (CASM)
    *   **Why it matters:** A key efficiency metric, showing the cost of flying one seat one mile. Lower CASM indicates better cost control.
    *   **Best chart:** Gauge chart or number card for the overall average, line chart for trend over time, bar chart for CASM by carrier or aircraft type.
    *   **Drill-down fields:** Carrier, Aircraft Type, Date/Month, Cost Category (e.g., fuel, labor).

### Create a Synthetic Airline Dataset (if you don't have one)
Use this dataset to prototype visualizations and dashboards. It contains realistic columns for the KPIs you defined. You can always swap in your BigQuery data later.


````python
# Generate a synthetic airline dataset
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

np.random.seed(467)

dates = pd.date_range(start='2024-01-01', end='2024-06-30', freq='D')
carriers = ['AA', 'DL', 'UA', 'WN', 'AS']
origins = ['ATL','DFW','DEN','ORD','LAX','SEA','SFO','PHX','CLT','MCO']
dests =   ['JFK','LGA','BOS','SFO','LAX','SEA','MIA','IAH','IAD','PHL']

rows = []
for d in dates:
    for carrier in carriers:
        # number of flights per day per carrier
        n = np.random.randint(80, 140)
        for _ in range(n):
            o = np.random.choice(origins)
            de = np.random.choice(dests)
            if o == de:
                continue
            distance = np.random.randint(300, 2800)
            seats = np.random.choice([90,120,150,180,210])
            load_factor = np.clip(np.random.normal(0.82, 0.07), 0.55, 0.98)
            pax = int(seats * load_factor)
            dep_delay = int(np.random.normal(8, 15))
            arr_delay = dep_delay + int(np.random.normal(0, 12))
            dep_delay = max(dep_delay, -5)
            arr_delay = max(arr_delay, -10)
            cancelled = np.random.rand() < 0.025
            diverted = np.random.rand() < 0.005
            revenue = pax * np.random.uniform(120, 420)
            casm = np.random.uniform(0.08, 0.16)  # cost per available seat mile
            cost = casm * seats * distance
            fuel_burn = distance * np.random.uniform(3.0, 4.5)  # gallons per seat-mile proxy
            nps = np.random.normal(35, 15)  # if you want a customer metric proxy
            rows.append({
                'date': d,
                'carrier': carrier,
                'origin': o,
                'dest': de,
                'distance_miles': distance,
                'seats': seats,
                'passengers': pax,
                'dep_delay_min': dep_delay,
                'arr_delay_min': arr_delay,
                'cancelled': int(cancelled),
                'diverted': int(diverted),
                'revenue_usd': round(revenue, 2),
                'cost_usd': round(cost, 2),
                'fuel_burn_gal': round(fuel_burn, 1),
                'nps': round(nps, 1)
            })

df_flights = pd.DataFrame(rows)

# Derived KPIs
df_flights['on_time'] = (df_flights['arr_delay_min'] <= 15).astype(int)
df_flights['load_factor'] = (df_flights['passengers'] / df_flights['seats']).round(3)
df_flights['yield_usd_per_pax_mile'] = (df_flights['revenue_usd'] / (df_flights['passengers'] * df_flights['distance_miles']).clip(lower=1)).round(4)
df_flights['asm'] = df_flights['seats'] * df_flights['distance_miles']  # available seat miles
df_flights['casm'] = (df_flights['cost_usd'] / df_flights['asm']).round(4)
df_flights['month'] = df_flights['date'].dt.to_period('M').astype(str)

df_flights.head()
````

````python
# Compute On-Time Rate by Carrier
on_time_by_carrier = df_flights.groupby('carrier')['on_time'].mean().sort_values(ascending=False) * 100

# Plotting
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=on_time_by_carrier.index, y=on_time_by_carrier.values, hue=on_time_by_carrier.index, palette='viridis', legend=False)


# Add labels and title
plt.xlabel('Carrier')
plt.ylabel('On-Time Rate (%)')
plt.title('On-Time Rate by Carrier')
plt.ylim(0, 100) # Set y-axis limit to 100%

# Add percentage labels on bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%')

plt.tight_layout()
plt.show()
````

````python
# Aggregate to daily average arrival delay
daily_avg_delay = df_flights.groupby('date')['arr_delay_min'].mean().reset_index()

# Calculate 7-day rolling average
daily_avg_delay['rolling_avg_delay'] = daily_avg_delay['arr_delay_min'].rolling(window=7).mean()

# Plotting
plt.figure(figsize=(12, 6))
sns.lineplot(data=daily_avg_delay, x='date', y='arr_delay_min', label='Daily Average Delay')
sns.lineplot(data=daily_avg_delay, x='date', y='rolling_avg_delay', color='red', linestyle='--', label='7-Day Rolling Average')

# Add axis labels, title, and grid
plt.xlabel('Date')
plt.ylabel('Average Arrival Delay (minutes)')
plt.title('Average Arrival Delay Over Time')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
````

````python
# Plotting the histogram of arrival delays
plt.figure(figsize=(10, 6))
sns.histplot(df_flights['arr_delay_min'], bins=50, kde=True)

# Add a vertical line at 0
plt.axvline(0, color='red', linestyle='--', label='0 Minutes Delay')

# Annotate percent of flights within [-5, 15] minutes
on_time_count = df_flights[(df_flights['arr_delay_min'] >= -5) & (df_flights['arr_delay_min'] <= 15)].shape[0]
total_flights = df_flights.shape[0]
on_time_percent = (on_time_count / total_flights) * 100

plt.text(15, plt.ylim()[1]*0.8, f'{on_time_percent:.1f}% within [-5, 15] min', ha='left', color='green')

# Add labels and title
plt.xlabel('Arrival Delay (minutes)')
plt.ylabel('Frequency')
plt.title('Distribution of Arrival Delays')
plt.legend()
plt.tight_layout()
plt.show()
````

````python
# Compute mean load factor by route (origin-dest pair)
df_flights['route'] = df_flights['origin'] + ' - ' + df_flights['dest']
route_load_factor = df_flights.groupby('route')['load_factor'].mean().sort_values(ascending=False).head(15)

# Plotting
plt.figure(figsize=(10, 8))
sns.barplot(x=route_load_factor.values, y=route_load_factor.index, hue=route_load_factor.index, palette='viridis', legend=False)

# Add labels and title
plt.xlabel('Average Load Factor')
plt.ylabel('Route')
plt.title('Top 15 Routes by Average Load Factor')

plt.tight_layout()
plt.show()
````

````python
# Plotting Cost vs. Revenue with CASM Coloring
plt.figure(figsize=(10, 8))
ax = sns.scatterplot(data=df_flights, x='cost_usd', y='revenue_usd', hue='casm', palette='viridis', alpha=0.6, s=50)

# Add a diagonal y=x reference line
max_val = max(df_flights['cost_usd'].max(), df_flights['revenue_usd'].max())
plt.plot([0, max_val], [0, max_val], color='red', linestyle='--', label='Cost = Revenue')

# Add labels and title
plt.xlabel('Cost (USD)')
plt.ylabel('Revenue (USD)')
plt.title('Cost vs. Revenue (Colored by CASM)')

# Add legend for CASM color scale
# The scatterplot already creates the colorbar based on 'hue'.
# We can get the colorbar artist from the axes if needed for further customization,
# but usually, seaborn handles it automatically when hue is a continuous variable.
# Let's remove the manual colorbar creation and rely on seaborn.

plt.legend() # Keep the legend for the Cost=Revenue line
plt.tight_layout()
plt.show()
````
